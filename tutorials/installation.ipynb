## Installation

This is an installtion using conda, for installation instructions using no environment or to compile using cmake see: https://github.com/hi-chi/pipic/blob/main/docs/guides/INSTALLATION.md. However, some of the extensions used in these tutorials might not be included in the installtion provided by ```pip install pipic```.

0. Download the repositiory or clone by ```git clone -b pipic_tutorial --single-branch git://github.com/hi-chi/pipic.git``` 

1. If not installed, install conda following the guide at https://www.anaconda.com/docs/getting-started/miniconda/install

2. Install dependencies: in terminal run
``` conda env create -f environment.yml ```

3. Choose the pipic_env kernel in jupyter notebook (for the new environment to show up ipykernel needs to be install in the same enviroment as jupyter notebook. To install ipykernel run ```pip install ipykernel``` in this environment).
